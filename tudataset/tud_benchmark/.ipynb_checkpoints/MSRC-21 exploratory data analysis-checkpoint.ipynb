```python
def create_img_index_by_class(classes):
    id2img = pd.read_csv(os.path.join(os.getcwd(),"datasets","MSRC_21","MSRC_21","raw","MSRC_21_map_id2im.csv"), sep='\t',header=None)
    file_names = np.array(id2img[1])
    converted_filenames = np.full_like(file_names,"")
    #print(converted_filenames.shape)
    for i in range(0,len(file_names)):
        #image = Image.open(os.path.join(os.getcwd(),"Images",file_names[i]))
        pattern = r'(.*)\.bmp'
        result = re.match(pattern, file_names[i]).group(1)
        new_filename = result+".png"
        #image.save("Images/"+ new_filename)
        converted_filenames[i] = new_filename
    file_classes = np.vstack([converted_filenames,classes]).T
    return file_classes
```

## MSRC_21 dataset

Old benchmark set for image identification
Here we use the MSRC_21 data set, but it works the same for all other data sets

node labels - semantic information on image parts available, we simplify the task to finding main theme


"The MSRC datasets are associated with computer vision tasks. Images are encoded by graphs, where vertices represent superpixels with a semantic label and edges their adjacency." 

Neumann 2015: "The nodes of each graph are derived by oversegmenting the images using the quick shift algo- rithm,6 resulting in one graph among the superpixels of each image. Nodes are connected if the superpixels are adjacent, and each node can further be annotated with a semantic label. Imagining an image retrieval system, where users provide images with semantic information, it is realistic to assume that this information is only available for parts of the images, as it is easier for a human annotator to label a small number of image regions rather than the full image. As the images in the msrc datasets are fully annotated, we can derive semantic (ground-truth) node labels by taking the mode ground-truth label of all pixels in the corre- sponding superpixel. Semantic labels are, for example, building, grass, tree, cow, sky, sheep, boat, face, car, bicycle, and a label void to handle objects that do not fall into one of these classes. "

Node label information contains noise, some nodes are unlabelled while other's are of the void category. 

Synthetic dataset with assumption that humans can quickly pre-label a segmented image.

```python

def cluster_average_nodes(G):
    nodes_per_class = dict.fromkeys(np.unique(classes),0)
    graphs_per_class = dict.fromkeys(np.unique(classes),0)

    for graph in G:
        cluster = graph.graph["classes"][0]
        nodes_per_class[cluster] += len(graph.nodes)
        graphs_per_class[cluster] += 1

    averages = {}

    for cluster,no_nodes in nodes_per_class.items():
        averages[cluster] = no_nodes / graphs_per_class[cluster]
    return averages
```

## Outlier detection

We use the Local Outlier Factor to check our KernelPCA Gram matrices for outliers and plot them against the first two dimensions. For clarity, only the large values of LOF are plotted.

```python
figs={}

# adapted from sci-kit learn example: https://scikit-learn.org/stable/auto_examples/neighbors/plot_lof_outlier_detection.html

for name,data in kpcas200_grams.items():
    clf = sklearn.neighbors.LocalOutlierFactor()
    clf.fit(data)
    X_scores = clf.negative_outlier_factor_
    radius = (X_scores.max() - X_scores) / (X_scores.max() - X_scores.min())
    radius[radius < 0.25] = 0
    figs[name] = plt.figure()
    


    axs[name] = figs[name].add_subplot(111)
    axs[name].set_title("LOF for {}".format(name))
    axs[name].scatter(data[:,0], data[:,1], c=classes, s=3.0, label="Data points")
    axs[name].scatter(
        data[:, 0],
        data[:, 1],
        s= 1000 * radius,
        edgecolors="b",
        facecolors="none",
        label="Outlier scores",
    )
    legend = axs[name].legend(loc="upper left")

```

The Graphlet and Shortest-path kernel contain significantly more / higher outliers than the WL kernels. For the WL kernel it is also notable that what appeared to be densely clustered points in the first two dimensions of the KernelPCA are actually far more distant from each other than appears.

```python
def plot_graphs_by_id(id_list,G, color = None):    
    '''
    Plot up to 5 images by id
    '''
    plt.figure(figsize=(7,7))
    if len(id_list) > 5:
        raise ValueError("Can only plot 5 images")
    fig, axes = plt.subplots(1,5)
    for i,graph_index in enumerate(id_list):
        axes[i] = nx.draw_networkx(G[graph_index], 
                     pos=nx.spring_layout(G[graph_index], seed=42),
                     with_labels=True,
                     node_color=color,
                     cmap="Set2")
        axes[i].axis('off')
    #for i in range(len(loaded_images),len(axes)):
    #    axes[i].set_visible(False)
    plt.title("Graphs with ids: {}".format(id_list))
```

```python
# only keep the first 10 components to speed up computation
reduced_kpca_10 = reduced_kpca[:, 0:10]
```

```python
costs = []
kmeans_runs = []
k_list = [1, 2, 5, 10, 15, 20, 25, 30, 35, 40]
for k in k_list:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(reduced_kpca_10)
    costs.append(kmeans.inertia_)
    kmeans_runs.append(deepcopy(kmeans))
```

```python
def scree_plot(k_list, costs):
    plt.figure(figsize=(10,5))
    plt.plot(k_list, costs, marker="o")
    plt.xlabel('Number of Clusters')
    plt.ylabel('KMeans loss')
    plt.title('Search for the elbow to determine the optimal k')
    plt.show();
scree_plot(k_list, costs)
```

```python
selected_idx = 5
kmeans = kmeans_runs[selected_idx]
df = pd.DataFrame(reduced_kpca_10[:,0:4])
df["cluster"] = kmeans.labels_
sns.pairplot(df, hue="cluster", diag_kind="hist", height=4)
plt.show();
```

```python
from sklearn.metrics import pairwise_distances
# select first 2 nearest neighbours in for each cluster

# these could be good to visualise to get an idea of clusters!

nr_nearest = 2
nearest_indices = []
for cluster_i in set(kmeans.labels_):
    mask = (kmeans.labels_ == cluster_i)
    selection = reduced_kpca_10[mask]
    print(f"number of data points in cluster {cluster_i}: {selection.shape[0]}")
    center_i = kmeans.cluster_centers_[cluster_i].reshape(1,-1)
    distances_i = pairwise_distances(center_i, reduced_kpca_10)
    nearest_indices.append(np.argsort(distances_i, )[0][0:nr_nearest])
    print(f"Nearest data points in cluster {cluster_i}: {nearest_indices[-1]}")
```
